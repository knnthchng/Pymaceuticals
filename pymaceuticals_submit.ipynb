### Analysis

- Of the four drug treatments studied in detail, mice treated with Capomulin and Ramicane show the lowest median tumor volumes at the end of their observations compared to those treated with Infubinol and Ceftamin. Looking at the total number of timepoints (observations) for the four drugs, as visualized in the bar plots, Capomulin and Ramicane have more observations, which shows that more mice survived to be observed further along the experiment's duration. Additionally, Capomulin and Ramicane both have the lowest variance and standard error scores of all the drugs tested.

- Subject i557 is an anomaly as her tumor volume initially fell, and then kept rising after timepoint 5. Two other mice who also have been treated with Capomulin show a reduction in tumor volume over the course of their treatment. While her tumor volume data is very unusual, her data does not fall beyond the lower or upper boundaries to be considered an outlier. 

- Looking at the data for average weight and average tumor volume, there is a strong positive correlation between mouse weight and tumor volume. A causal relationship cannot be inferred from the data, especially if weight is a factor in tumor development and volume, though it may be possible that tumorous tissues can contribute to body mass.

```python
# Dependencies and Setup
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_path = "data/Mouse_metadata.csv"
study_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_path)
study_results = pd.read_csv(study_path)

# Combine the data into a single dataset
alldata = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
alldata
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
avg_tumor = alldata.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
med_tumor = alldata.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
var_tumor = alldata.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
dev_tumor = alldata.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
SEM_tumor = alldata.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

stats_tumor = pd.DataFrame({"Mean Tumor Volume":avg_tumor,
                           "Median Tumor Volume":med_tumor,
                           "Tumor Volume Variance":var_tumor,
                           "Tumor Volume Std. Dev.":dev_tumor,
                           "Tumor Volume Std. Err.":SEM_tumor})
stats_tumor
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
# https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/
stats_agg = alldata.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":
                                                ['mean', 'median', 
                                                'var','std', 'sem']})
stats_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_time = alldata.groupby("Mouse ID")["Timepoint"].max()
final_time = final_time.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_time_df = pd.merge(alldata, final_time, on=["Mouse ID","Timepoint"],
                         how="right")
final_time_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens = ['Capomulin','Ramicane','Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
final_tumorvol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in regimens:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumorvol = final_time_df.loc[final_time_df["Drug Regimen"] == drug, "Tumor Volume (mm3)"]
    
    # add subset 
    final_tumorvol.append(tumorvol)
    
    # Determine outliers using upper and lower bounds
    quartiles = round(tumorvol.quantile([0.25, 0.5, 0.75]),6)
    lowerq = round(quartiles[0.25],6)
    upperq = round(quartiles[0.75],6)
    iqr = round(upperq - lowerq,6)
    
    lower_bound = round(lowerq - (1.5*iqr),6)
    upper_bound = round(upperq + (1.5*iqr),6)
    outliers = tumorvol.loc[(tumorvol < lower_bound) | (tumorvol > upper_bound)]
    
    print(f"{drug} IQR data:")
    print(f"The lower quartile of {drug} is: {lowerq}")
    print(f"The upper quartile of {drug} is: {upperq}")
    print(f"The interquartile range of {drug} is: {iqr}")
    print(f"The median of {drug} is: {quartiles[0.5]}")
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    print("--------------------------------------------------------")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.figure(figsize=(10,8))
plt.title("Tumor Volume for Drug Regimens", fontsize=16)
plt.xlabel("Drug Regimen", fontsize=14)
plt.ylabel("Final Tumor Volume (mm3)",fontsize=14)
plt.boxplot(final_tumorvol,labels=regimens,showmeans=True)

plt.savefig("Images/Final_Tumor_Volume_Boxplot.png",facecolor="white")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capom_data = alldata.loc[alldata["Drug Regimen"] == "Capomulin"]
capom_ID = capom_data["Mouse ID"].unique()
# Selected subject i557
sub_i557 = capom_data.loc[capom_data["Mouse ID"] == "i557"]
sub_i557
```

```python
sub_s185 = capom_data.loc[capom_data["Mouse ID"] == "s185"]
sub_x401 = capom_data.loc[capom_data["Mouse ID"] == "x401"]
```

```python
x_i557 = sub_i557["Timepoint"]
vol_i557 = sub_i557["Tumor Volume (mm3)"]
x_s185 = sub_s185["Timepoint"]
vol_s185 = sub_s185["Tumor Volume (mm3)"]
x_x401 = sub_x401["Timepoint"]
vol_x401 = sub_x401["Tumor Volume (mm3)"]

plt.figure(figsize=(15,10))
plt.plot(x_i557, vol_i557, marker='s',color='g')
plt.plot(x_s185, vol_s185, marker='o',color='b')
plt.plot(x_x401, vol_x401, marker='v',color='r')
plt.legend(['i557','s185','x401'], loc='best')
plt.ylim(20, 50)
plt.title("Treatment under Capomulin", fontsize=16)
plt.xlabel("Timepoint (Days)", fontsize=14)
plt.ylabel("Tumor Volume (mm3)", fontsize=14)

plt.savefig("Images/Subject_i557_Treatment.png",facecolor='white')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capom_avg_vol = capom_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
capom_avg_wt = capom_data.groupby("Mouse ID")["Weight (g)"].mean()

plt.figure(figsize=(10,8))
plt.title("Average Tumor Volume vs. Average Weight for Mice under Capomulin")
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.scatter(capom_avg_wt, capom_avg_vol)

plt.savefig("Images/Average_Weight_and_Tumor_Vol.png",facecolor="white")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
# This cell calculates the correlation coefficient
correl_volwt = st.pearsonr(capom_avg_wt,capom_avg_vol)

print(f"The correlation coefficient between average tumor volume")
print(f"and average weight for mice taking capomulin is {round(st.pearsonr(capom_avg_wt,capom_avg_vol)[0],2)}.")
```

```python
# Calculate the linear regression model and display it in a scatterplot:
cap_slope, cap_int, cap_r, cap_p, cap_std_err = st.linregress(capom_avg_wt,capom_avg_vol)
cap_lin = (cap_slope * capom_avg_wt) + cap_int # Equation for linear regression model

plt.figure(figsize=(10,8))
plt.title("Average Tumor Volume vs. Average Weight for Mice under Capomulin")
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.scatter(capom_avg_wt, capom_avg_vol)
plt.plot(capom_avg_wt, cap_lin, color='r')

plt.savefig("Images/Average_Weight_and_Tumor_LINEAR.png",facecolor="white")
plt.show()
```
